```python
import pandas as pd
import numpy as np
import cv2
import os

# Read the CSV file
csv_file = r"i:\Dataset\annotations_handheld.csv"
data = pd.read_csv(csv_file)

# Output folder for masks and masked images
output_folder = r"i:\Dataset\output_masks"

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Iterate over each row
for index, row in data.iterrows():
    image_path = row['image']
    x1, y1, x2, y2 = row['x1'], row['y1'], row['x2'], row['y2']
    
    # Ensure coordinates are valid
    if x2 <= x1 or y2 <= y1:
        print(f"Invalid coordinates for image: {image_path}")
        continue
    
    # Create a mask
    mask = np.zeros((y2 - y1, x2 - x1), dtype=np.uint8)
    mask[:y2 - y1, :x2 - x1] = 255  # Set the region defined by (x1, y1) and (x2, y2) to 255
    
    # Save the mask
    mask_filename = os.path.join(output_folder, f'{os.path.splitext(os.path.basename(image_path))[0]}_mask.png')
    cv2.imwrite(mask_filename, mask)

    print(f'Mask saved for {image_path}')
```

```python
import numpy as np
import pandas as pd

# Function to generate masks from bounding box coordinates
def generate_masks(csv_file):
    # Read CSV file
    df = pd.read_csv(csv_file)
    
    for index, row in df.iterrows():
        x1, y1, x2, y2 = row['x1'], row['y1'], row['x2'], row['y2']
        
        # Create mask
        mask = np.zeros((y2, x2), dtype=np.uint8)
        mask[y1:y2, x1:x2] = 255
        
        # Save mask
        mask_name = f"mask_{index}.png"
        cv2.imwrite(mask_name, mask)

# Example usage
csv_file = r"I:\Dataset\annotations_handheld.csv"
generate_masks(csv_file)
```
